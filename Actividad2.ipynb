```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_csv('Topo.TXT',skiprows=4,sep='\s+')
```

```python
df = pd.DataFrame(df0)
```

```python
df.dtypes
```

```python
df['FECHA'] = pd.to_datetime(df.apply(lambda x: x['DD/MM/AAAA'] + ' ' + x['HH:MM'], 1), dayfirst=True)
df = df.drop(['DD/MM/AAAA', 'HH:MM'], 1)
```

```python
plt.figure(); df.TEMP.plot(); plt.legend(loc='best')
plt.title("Temperatura")
plt.ylabel("Grados CentÃ­grados")
plt.grid(True)
plt.show()
```

```python
df1 = df[['TEMP','HR']]
plt.figure(); df1.plot(); plt.legend(loc='best')
plt.title("VariaciÃ³n de la Temperatura y la Humedad Relativa")
plt.ylabel("Temp ÂºC /(%) HR")
plt.grid(True)
plt.show()
```

```python
plt.plot_date(x=df.FECHA, y=df.TEMP, fmt="b-")
plt.title("VariaciÃ³n de la Temperatura")
plt.ylabel("Temp ÂºC")
plt.grid(True)
plt.show()
```

```python
rv = df[["VELS","VELR"]]
plt.plot(df.FECHA,rv)
plt.title("Velocidad de los vientos y de las rÃ¡gas a travÃ©s del tiempo")
plt.grid(True)
plt.show
```
